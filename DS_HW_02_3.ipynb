**Пошук та сортування даних**

```python
# Який рейтинг у датасеті найвищий? Відповідь:

max_raiting = df['user_rating'].max()
print(max_raiting)
```

```python
# Скільки книг мають такий рейтинг? Відповідь:

books_with_max_raiting = df[df['user_rating'] == max_raiting]
quantity_of_books = len (books_with_max_raiting['name'].unique())
print(quantity_of_books)
```

```python
# Яка книга має найбільше відгуків? Відповідь:

book_with_max_review = df[df['reviews'] == df['reviews'].max()]
max_review_book_name = book_with_max_review['name'].values[0]
print (max_review_book_name)
```

```python
# Відповідь: З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча (можна використати проміжний датафрейм)? Відповідь:

top_50_books = df.nlargest(50, 'user_rating')
more_expensive_from_50 = top_50_books[top_50_books['price'] == top_50_books['price'].max()]['name'].values[0]
print (more_expensive_from_50)
```

```python
# Відповідь: Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році (використовуйте &)? Відповідь:

top_50_fiction = df[(df['genre'] == 'Fiction')& (df['year'] == 2010)].nlargest(50, 'user_rating')
print (len(top_50_fiction))
```

```python
# Відповідь: Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)? Відповідь:

books_4_9_2010_2011 = df[(df['user_rating'] == 4.9) & (df['year'].isin([2010, 2011]))]
number_of_books = len(books_4_9_2010_2011)
print(number_of_books)
```

```python
# І насамкінець, давайте відсортуємо за зростанням ціни всі книги, які потрапили до рейтингу в 2015 році і коштують дешевше за 8 доларів (використовуйте функцію sort_values).

books_2015_under_8 = df[(df['year'] == 2015) & (df['price'] < 8)]
sorted_books = books_2015_under_8.sort_values(by='price', ascending=True)

sorted_books.tail()
```

```python
# Відповідь: Яка книга остання у відсортованому списку? Відповідь:

last_book = sorted_books.tail(1)
print(last_book['name'].values[0])
```

```python
# Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count). Не беріть усі стовпці, виберете тільки потрібні

author_books_count = df.groupby('author').agg({'name': 'count'}).reset_index()
author_books_count = author_books_count.rename(columns={'name': 'quantity'})
author_books_count.head()
```

```python
#Відповідь: Якої розмірності вийшла таблиця? Відповідь:

author_books_count.shape
```

```python
# Відповідь: Який автор має найбільше книг? Відповідь:

author_max_books = author_books_count[author_books_count['quantity'] == author_books_count['quantity'].max()]['author'].values[0]
print(author_max_books)
```

```python
#Відповідь: Скільки книг цього автора? Відповідь:

quantity = author_books_count[author_books_count['author'] == 'Jeff Kinney']['quantity'].values[0]
print(quantity)
```

```python
# Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні

author_avg_rating = df.groupby('author').agg({'user_rating': 'mean'}).reset_index()
author_avg_rating.head()
```

```python
# Відповідь: У якого автора середній рейтинг мінімальний? Відповідь:

author_min_avg_rating = author_avg_rating[author_avg_rating['user_rating'] == author_avg_rating['user_rating'].min()]['author'].values[0]
print(author_min_avg_rating)
```

```python
# Відповідь: Який у цього автора середній рейтинг? Відповідь:

avg_rating = author_avg_rating[author_avg_rating['author'] == 'Donna Tartt']['user_rating'].values[0]
print(avg_rating)
```

```python
# З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну

merged_df = pd.merge(author_books_count, author_avg_rating, on='author', how='outer')
merged_df.head()
```

```python
# Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)

sorted_merged_df = merged_df.sort_values(by=['quantity', 'user_rating'], ascending=[True, True])
sorted_merged_df.head()
```

```python
# Відповідь: Який автор перший у списку?

first_author = sorted_merged_df['author'].iloc[0]
print(first_author)
```
